# Feature Engineering

Date Features

```python
# Date Features
features=df
features['Year']=df['date'].dt.year
features['Month']=df['date'].dt.month
features['Day']=df['date'].dt.day
features
```

Lag Features

```python
#getting value of previous day birth
features['lag1']=features['births'].shift(1)
#getting value of same day previous year
features['lag2']=features['births'].shift(365)

features
```

Window Features

```python
#mean of current and last day value
features['Roll_mean']=features['births'].rolling(window=2).mean()
#Max of current and last 2 days value
features['Roll_max']=features['births'].rolling(window=3).max()

features
```

Expanding Features


```python
#Expanding is similar to Rolling but it is withour window
#Max. birth value till the record

features['Expand_max']=features['births'].expanding().max()
features
```
